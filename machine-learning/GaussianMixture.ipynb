# Experiment - 9
### Aim: To study and implement Gaussian Mixture Model using the Expectation Maximization.
### Abstract:
The objective of this experiment is to study and implement Gaussian Mixture Model (GMM) using the Expectation Maximization (EM) algorithm. GMM is a probabilistic model that represents the distribution of data points as a mixture of Gaussian components. The EM algorithm iteratively estimates the parameters of the GMM to maximize the likelihood of the observed data. In this experiment, we aim to explore the performance of GMM in capturing complex data distributions, understand the EM algorithm, evaluate model fit using information criteria, and visualize the results using plots in a Jupyter Notebook.

### Procedure:
#### Step 1. Importing Necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.mixture import GaussianMixture
```

#### Step 2: Generating synthetic data using make_blobs

```python
X, y = make_blobs(n_samples=1000, centers=4, n_features=2, random_state=42)
print("X",X[:5])
print("y",y[:5])
```

#### Step 3: Initializing and fitting the Gaussian Mixture Model using Expectation Maximization

```python
gmm = GaussianMixture(n_components=4, random_state=42)
gmm.fit(X)
```

#### Step 4: Predicting the cluster labels for the data points

```python
labels = gmm.predict(X)
print(labels[:5], len(labels))
```

#### Step 5: Visualizing the data points and estimated GMM components

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=labels, palette='viridis')
plt.scatter(gmm.means_[:, 0], gmm.means_[:, 1], c='red', marker='x', s=100)
plt.title("Gaussian Mixture Model")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

#### Step 6: Calculating the AIC and BIC scores

```python
aic = gmm.aic(X)
bic = gmm.bic(X)
print("AIC:", aic)
print("BIC:", bic)
```

### Conclusion: 
By conducting this experiment, we gained a better understanding of Gaussian Mixture Model and the Expectation Maximization algorithm and explored visualization techniques to interpret the learned data distribution effectively.
