# Experiment - 8
### Aim: To study and implement K Means Clustering to find natural patterns in data.
### Abstract:
The objective of this experiment is to study and implement K-means clustering, a popular unsupervised learning algorithm, to identify natural patterns in data. K-means clustering is a centroid-based algorithm that groups data points into clusters based on their proximity. In this experiment, we will explore the performance of K-means clustering on a selected dataset, determine the optimal number of clusters, evaluate clustering quality metrics, and visualize the results using plots in a Jupyter Notebook.

### Procedure:
#### Step 1. Importing Necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

#### Step 2: Generating synthetic data using make_blobs

```python
X, y = make_blobs(n_samples=1000, centers=4, n_features=2, random_state=42)
print("X",X[:5])
print("y",y[:5])
```

#### Step 3: Initializing and fitting the K-means clustering algorithm

```python
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)
```

#### Step 4: Predicting the cluster labels for the data points

```python
labels = kmeans.predict(X)
print(labels[:5], len(labels))
```

#### Step 5: Visualizing the clusters

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=labels, palette='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x', s=100)
plt.title("K-means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

#### Step 6: Calculating the silhouette score

```python
silhouette_avg = silhouette_score(X, labels)
print("Silhouette Score:", silhouette_avg)
```

### Conclusion: 
By conducting this experiment, gained a better understanding of K-means clustering and explored visualization techniques to interpret the discovered natural patterns effectively.
